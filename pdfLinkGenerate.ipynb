```python
import pandas as pd
import re
```

```python
data = pd.read_excel('vbpq.xlsx')
data
```

```python
data = data.drop(columns=['ng-binding 2','ng-binding 5', 'ng-binding 9'])
```

```python
data
```

```python
def extract_id(url):
    pattern = r'id=(\d+)'
    match = re.search(pattern, url)
    if match:
        return match.group(1)
    else:
        return None
```

```python
data['extracted_id'] = data['panel-heading href'].apply(extract_id)
```

```python
data = data.drop(columns=['panel-heading href'])
```

```python
output_filename = "extracted_id.csv"
data.to_csv(output_filename, index=False)
```

```python
import csv
```

```python
data
```

```python
data['extracted_id'] = pd.to_numeric(data['extracted_id'], errors='coerce')
data['extracted_id'] = data['extracted_id'] + 1
```

```python
ids = data['extracted_id'].to_list()
```

```python
base_url = "https://vbpq.mof.gov.vn/DKC.FileManagement/FileStorage/File/{}"
```

```python
urls = [base_url.format(id) for id in ids]
```

```python
csv_filename = "generated_links.csv"
with open(csv_filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['URL'])  
    for url in urls:
        writer.writerow([url])
```
